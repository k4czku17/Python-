```python
import matplotlib, matplotlib, pandas #importuje biblioteki 

df = pandas.read_csv('lot.csv', encoding= 'unicode_escape') #wczytuje plik csv, dotyczacy lotow w latach od 1958 do 1960
df.head() #wyświetla czołową część pliku
```

```python
df['1958'] #wyświetla dane z 1958
```

```python
df['1959']
```

```python
df[['Month','1960']] #wyświetla 2 rekordy z pliku 
```

```python
df[['1958','1959','1960']]
```

```python
x=df['Month']
y=df['1958']
z=df['1959']
a=df['1960']
#przypisuje dane 
```

```python
import pylab 

pylab.plot(x, y) #dane ktore maja byc przedstawione na wykresie 
pylab.title('Miesieczne loty w 1958') #dodanie tytulu 
pylab.grid(True)
pylab.show() 
```

```python
fig, ax =  fig, ax = matplotlib.pyplot.subplots(figsize=(15,9)) #ustala rozmiat figury 
ax.pie([4590, 5142, 5714], labels=['1958', '1959', '1960']) #dane ktore beda mialy byc wyswietlone 
ax.legend() #dodanie legendy 
pylab.title('Zestawienie ze sobą 3 lat') 
```

```python
fig, ax = matplotlib.pyplot.subplots(figsize=(15,9))
ax.bar(x,z)
ax.bar(x,y)
pylab.title('Porównanie miesięczne pomiedzy 1958r a 1959r.')
```

```python
fig, ax = matplotlib.pyplot.subplots(figsize=(15,9))
ax.scatter(x, y, marker='x', alpha=1.0)
ax.scatter(x, z, marker='s', alpha=1.0)
ax.scatter(x, a, marker='d', alpha=1.0)
pylab.grid(True)
ax.legend(['1958','1959','1960'])
pylab.title('Porównanie 3 lat')
```

```python
matplotlib.pyplot.style.use('dark_background') #ustawienie ciemnego tła 
fig, ax = matplotlib.pyplot.subplots(figsize=(15,9))
res = ax.plot(x, y)
res = ax.plot(x, z)
res = ax.plot(x, a)
ax.legend(['1958','1959','1960'])
pylab.title('Zestawienie 3 lat')
```
